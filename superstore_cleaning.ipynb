# when order date column contains multuple formats

```python
import datetime
```

```python
# Clean date strings: remove spaces, convert to string, strip trailing whitespace
df['order_date'] = df['order_date'].str.strip().str.replace('-', '/')
```

```python
# Final parsing function
def parse_to_dmy(date_str):
    try:
        return datetime.strptime(date_str, '%d/%m/%Y').strftime('%d/%m/%Y')
    except:
        try:
            return datetime.strptime(date_str, '%m/%d/%Y').strftime('%d/%m/%Y')
        except:
            return None
```

```python
# Apply it
df['order_date'] = df['order_date'].apply(parse_to_dmy)
```
